### Разрыв между качеством данных активных и неактивных пользователей очень велик. 

Моей основной задачей является создание рекомендательной системы, которая глубже вовлечет пользователей в экосистему банка, поэтому неактивными пользователями пренебрегаем. В результате количество строк в датасете уменьшится с 13 миллионов до 6 миллионов.

```python
# Какие есть продукты

product_cols = [
    col for col in df_active.columns if col.startswith("ind_") and col.endswith("_ult1")
]

print(product_cols)
```

```python
df_active[product_cols] = df_active[product_cols].fillna(0).astype(int)
```

```python
df_active[product_cols].describe()
```

```python
del df
```

```python
def fill_missing_values(df:pd.DataFrame,product_cols:list)->pd.DataFrame:
    for col in df.columns:
        if isinstance(df_active[col].mode()[0], (int,float)) and col not in(product_cols):
            df[col] = pd.to_numeric(df[col], errors='coerce')
            df[col] = df[col].fillna(df[col].mean())
        else :
            df[col] = df[col].fillna(df[col].mode().iloc[0])
    return df
```

```python
df_active_fill_miss = fill_missing_values(df_active,product_cols=product_cols)
```

```python
for col in df_active_fill_miss.columns:
    n_unique = df_active_fill_miss[col].nunique(dropna=True)
    n_nulls = df_active_fill_miss[col].isnull().sum()
    if n_unique <= 4 and n_nulls > 0:
        print(f"{col}: уникальных значений = {n_unique}, пропусков = {n_nulls}, уникальные значения: {df_active_fill_miss[col].unique()}")
```

### Разделяем выборки на тестовую и тренировочную

```python
from utils import split_data

train, test = split_data(ratio=0.75,data=df_active_fill_miss)
```

```python
sum_products_by_month = train.groupby(['fecha_dato']).agg(dict(zip(product_cols, ['sum'] * len(product_cols)))).reset_index()
```

```python
from pathlib import Path

graphs_dir = Path()/'graphs'
graphs_dir.mkdir(exist_ok=True)
```

### Частота владения продуктами 

```python
import seaborn as sns
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows=13, ncols=1, figsize=(10, 70))

months = sum_products_by_month['fecha_dato'].unique()

for i, month in enumerate(months):
    month_data = sum_products_by_month.loc[
        sum_products_by_month['fecha_dato'] == month, product_cols
    ]
    month_data.T.plot(kind='barh', legend=False, ax=axs[i])
    axs[i].set_title(f"Месяц: {month}")
    axs[i].set_xlabel("Продажи")
    axs[i].set_ylabel("Продукт")

plt.tight_layout()
plt.savefig(graphs_dir/"product_sales_by_month")
plt.show()

```
